# DEMO ON PLOTTING WITH MATHPLOTLIB

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
time = (0, 1, 2, 4)
position = (0, 10, 50, 100)
```

```python
plt.plot(time, position)
plt.xlabel('dopa')
```

```python
df = pd.read_csv('gapminder_gdp_africa.csv', index_col='country')
algeria_gdp = df.loc['Algeria',:]
nigeria_gdp = df.loc['Nigeria',:]
uganda_gdp = df.loc['Uganda',:]
cameroon_gdp = df.loc['Cameroon',:]
south_africa_gdp = df.loc['South Africa',:]
years = df.columns

plt.plot(years, algeria_gdp, label = 'Algeria')
plt.plot(years, nigeria_gdp, label = 'Nigeria')
plt.plot(years, uganda_gdp, label = 'Uganda')
plt.plot(years, cameroon_gdp, label = 'Cameroon')
plt.plot(years, south_africa_gdp, label = 'South Africa')
plt.legend()
plt.savefig('african_gpd.png')
```
